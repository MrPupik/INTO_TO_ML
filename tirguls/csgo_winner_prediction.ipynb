```python
ct_money_mean = data['ct_money'].mean()
# ata[data['ct_helmets']]['ct_helmets']
rich_helmet_median = data[data['ct_money']>=ct_money_mean]['ct_helmets'].median()
poor_helmet_median = data[data['ct_money']<ct_money_mean]['ct_helmets'].median()
```

```python
data.loc[data['ct_money']>=ct_money_mean,'ct_helmets'] = data.loc[data['ct_money']>=ct_money_mean,'ct_helmets'].fillna(rich_helmet_median)
data.loc[data['ct_money']<ct_money_mean,'ct_helmets'] = data.loc[data['ct_money']<ct_money_mean,'ct_helmets'].fillna(poor_helmet_median)
```

```python
a = data.isnull().sum()
a[a>0]
```

```python
ct_utility_cols = [x for x in data.columns if x.startswith('ct_grenade')]
t_utility_cols = [x for x in data.columns if x.startswith('t_grenade')]
data['t_utility'] = data[t_utility_cols].sum(axis=1)
data['ct_utility'] = data[ct_utility_cols].sum(axis=1)
```

```python
rifles = ['ak47','aug','awp','m4a4','famas','galilar','g3sg1','sg553','scar20']
ct_rifles = [f'ct_weapon_{rifle}' for rifle in rifles]
t_rifles = [f't_weapon_{rifle}' for rifle in rifles]
data['ct_rifles'] = data[ct_rifles].sum(axis=1)
data['t_rifles'] = data[t_rifles].sum(axis=1)
```

```python
t_weapons = [x for x in data.columns if x.startswith('t_weapon')]
ct_weapons = [x for x in data.columns if x.startswith('ct_weapon')]
all_weapons = t_weapons+ ct_weapons
all_util = t_utility_cols + ct_utility_cols
all_remove = all_weapons + all_util
df = data.drop(columns = all_remove,inplace=False)
df.shape
```

```python
df['round_winner'] = df['round_winner'].replace({'T':0,'CT':1})
df['bomb_planted'] = df['bomb_planted'].astype(int)
```

```python
corr = df.corr()
sns.heatmap(corr)
```

```python
df_proccesed = pd.get_dummies(df,prefix=['map'], columns = ['map'], drop_first=True)
ordered_cols = [x for x in df_proccesed.columns if x not in['round_winner']] + ['round_winner']
df_proccesed = df_proccesed[ordered_cols]
df_proccesed.shape
```

```python
X = df_proccesed.iloc[:,:-1].values
y = df_proccesed.iloc[:,-1]
y.shape
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.8,random_state=42)
```

```python
from sklearn.metrics import auc
from sklearn.metrics import RocCurveDisplay, RocCurveDisplay
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
clf = LogisticRegression()
param_grid = {
    'penalty': ['l1','l2'],
    'C': [0.1,1,10],
    'solver': ['saga']
}
optimizer = GridSearchCV(clf, param_grid,scoring='roc_auc',n_jobs=4)
```

```python
optimizer.fit(X_train,y_train)
```

```python
RocCurveDisplay.from_estimator(optimizer, X_test, y_test)
```
