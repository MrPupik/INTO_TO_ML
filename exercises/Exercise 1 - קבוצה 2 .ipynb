# 1. For the Artistic Value Of It
Find all movies that were **released before 2014**, have a **runtime above median runtime**, and made **revenue above 100M.**<br>
Show **top 7 by number of votes**.

```python
med = df['runtime'].median() 
df[(df.year < 2014) & (df.runtime > med) & (df.revenue_millions > 100)].sort_values(by=['votes'], ascending=False).head(7)
```

### 2. Best In The Industry
Find "highest-revenue" directors -<br>
Show **top 3 directors** (3 rows) that have the average highest revenue with more than 1 film.
Show what is the **average revenue** (column named `average_revenue_millions`) for each of them. <br>
show **how many films** (column named `n_films`) each one directed.<br>

```python
df5 = df.groupby('director').agg(
    n_films=('id', 'count'), average_revenue_millions=('revenue_millions', 'mean'))
df6 = df5[df5['n_films'] >1]
result = df6.sort_values('average_revenue_millions', ascending=[False]).head(3)
result
```

### 3. Best of the Best
Create and Show a table of the **highest rated film** by each of the 3 directors from ex.2 (the table will contain directors name, title, and rating).
<br>

```python
df_3 = df.loc[df['director'].isin(result.index.values)]
df_new = df_3[df_3.groupby('director')['rating'].transform('max')== df_3['rating']]
df_new[['director','title','rating']].sort_values('rating', ascending=[False])
```

### 5. Best actor
Create and show a table of the average revenue for the following actors: 
1. Leonardo DiCaprio
2. Matthew McConaughey
3. Johnny Depp

Show the table by descending order of average revenue

```python
df_5 = df[['actors', 'revenue_millions']].copy()
df_5.actors = df_5.actors.apply(lambda actors: [actor.strip() for actor in actors.split(',')])
df_new = df_5.explode('actors')
df_new = df_new.groupby('actors').mean().rename(columns={'revenue_millions': 'average_revenue'})
df_new = df_new.loc[df_new.index.isin(['Leonardo DiCaprio','Matthew McConaughey','Johnny Depp'])].sort_values('average_revenue',ascending=False)
df_new
```

B. Now plot the same plot again with log scale (y axis)
* name X axis - 'Revenue'
* name Y axis - 'Log Frequency'
* add the line for the median

```python
plt.hist(df['revenue_millions'], bins=50)
plt.gca().axvline(df['revenue_millions'].median(), color='r', linestyle='--')
plt.yscale("log")
plt.xlabel("Revenue")
plt.ylabel("Log Frequency")
plt.show()
```

### 7. Let's Go To The Movies
Plot the mean revenue per movie as line
